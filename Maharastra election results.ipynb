```python
from matplotlib import pyplot as plt
```

```python
u=df["win"].unique()
```

```python
c=["blue","black","brown","green","orange","red","purple","silver","white","yellow","tan","lime","khaki","cyan","darkgreen","pink"]
```

```python
fig,ax=plt.subplots(figsize=(35,15))
geodata.plot(column='city', scheme="quantiles",ax=ax,legend=True,cmap='coolwarm')
plt.xlim(70,85)
#plt.ylim(20,55)

for i in range(len(geodata)):
    plt.text(geodata.x[i],geodata.y[i],"{}\n{}".format(geodata.city[i],geodata.win[i]),size=10)
plt.title('MAHARASTRA ELECTION RESULTS 2019',fontsize=25)
plt.show()
```

```python
import matplotlib.patches as mpatches
```

```python
 patchList=[]
```

```python
fig,axis=plt.subplots(figsize=(40,15))
geodata.plot(ax=axis,legend=True,scheme="quantiles")
plt.xlim(70,85)
#plt.ylim(5,25)
# add countries names and numbers 
for i in range(len(u)):
    geodata[geodata["win"]==u[i]].plot(ax=axis,color=c[i],label=u[i],edgecolor='k',k=5)

for i in range(len(u)):
        data_key = mpatches.Patch(color=c[i], label=u[i])
        patchList.append(data_key)
    
    
plt.legend(handles=patchList)

plt.title('MAHARASTRA ELECTION RESULTS 2019',fontsize=25)

plt.show()
```
